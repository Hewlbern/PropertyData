```python
def Swing(States, url_swing):

    """
    Gets Swinging Property Datasets and turns it into a CSV format.
    """
    
    timestr = time.strftime("%Y%m%d-%H%M%S")
    Name= "swing_data_"+timestr+".csv"

    api_results = {}
    for items in States:
            url = url_swing + items
            request = requests.get(url)
            api_results[items]= pd.DataFrame(request.json())


    df = pd.concat(api_results).reset_index(level=1, drop=True).rename_axis('key').reset_index().set_index(['key'])
    df.to_csv(Name, sep='\t', encoding='utf-8')
    return df
```
